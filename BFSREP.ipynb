```python
Start: queue[5]
visited: [5]

step1: remove [5]
queue: [3,7]
visited: [5,3,7]

step2: remove [3]
queue: [7,2,4]
visited: [5,3,7,2,4]

step3: remove [7]
queue: [2,4,8]
visited: [5,3,7,2,4,8]

step4: remove [4]
queue: [4,8]

step5: remove [4]
neighobor 8 already visited
queue: [8]

Step 6:
Remove 8 -- no neighbors
Queue = []


Final output: 5,3,7,2,4,8
```

```python
# Graph
graph = {
    '5': ['3', '7'],
    '3': ['2', '4'],
    '7': ['8'],
    '2': [],
    '4': ['8'],
    '8': []
}

visited = []  # List for visited nodes
queue = []    # Initialize queue

def bfs(visited, graph, node):
    visited.append(node)
    queue.append(node)

    while queue:
        m = queue.pop(0)
        print(m, end=" ")

        for neighbour in graph[m]:
            if neighbour not in visited:
                visited.append(neighbour)
                queue.append(neighbour)

print("Following is the Breadth-First Search:")
bfs(visited, graph, '5')
```
