# DFS

```python
Start at 5
Visited: 5
Go to first neighbor → 3

At 3
Visited: 5, 3
Go to → 2

At 2
Visited: 5, 3, 2
No neighbors → backtrack to 3

Back to 3
Next neighbor → 4
Visited: 5, 3, 2, 4

At 4
Go to → 8
visited: 5, 3, 2, 4, 8
No neighbors → backtrack

Back to 5
Next neighbor → 7
Visited: 5, 3, 2, 4, 8, 7

At 7
Neighbor → 8 (already visited  skip)

output: 5 3 2 4 8 7
```

```python
graph = {
    '5': ['3','7'],
    '3': ['2','4'],
    '7': ['8'],
    '2': [],
    '4': ['8'],
    '8': []
}

visited = set()

def dfs(node):
    if node not in visited:
        print(node, end=" ")
        visited.add(node)
        for neighbour in graph[node]:
            dfs(neighbour)

print("Depth First Search:")
dfs('5')
```

# DFS Traversel

```python
Step 1: Start at 2

Visited: 2
Output: 2

Neighbors → 0, 3
Go to 0


Step 2: At 0

Visited: 2, 0
Output: 2 0

Neighbors → 1, 2
Go to 1

Step 3: At 1

Visited: 2, 0, 1
Output: 2 0 1

Neighbor → 2 already visited → skip
Backtrack to 0


Back to 0

Next neighbor → 2  skip
Backtrack to 2

Step 4: Back to 2

Next neighbor → 3

Visited: 2, 0, 1, 3


Output: 2 0 1 3

```

```python
from collections import defaultdict

class Graph:
    def __init__(self):
        self.graph = defaultdict(list)

    def addEdge(self, u, v):
        self.graph[u].append(v)

    def DFSUtil(self, v, visited):
        visited[v] = True
        print(v, end=" ")

        for i in self.graph[v]:
            if not visited[i]:
                self.DFSUtil(i, visited)

    def DFS(self, start):
        visited = [False] * 4  # 4 nodes: 0,1,2,3
        self.DFSUtil(start, visited)


# Driver code
g = Graph()
g.addEdge(0, 1)
g.addEdge(0, 2)
g.addEdge(1, 2)
g.addEdge(2, 0)
g.addEdge(2, 3)
g.addEdge(3, 3)

print("DFS starting from 2:")
g.DFS(2)
```
